```python
class A:
    def __init__(self, x, y, z):
        self.varX = x       
        self._varY = y        
        self._varZ = z    
    
    def demo(self):
        return f"ESTOY en A: x: {self.varX} -- y:{self._varY} --- z:{self.__varZ}"
class B(A):
    def __init__(self):
        A.__init__(self, "x", "y", "z")
            
    def demo(self):
        return f"ESTOY en B: x: {self.varX} -- y:{self._varY} --- z:{self._varZ}"
objB = B()
print(objB.demo())
```
