import pandas as pd

def obtener_respuestas_cdo(df, orden=None):
    """
    Retorna información de usuarios CDO que completaron la encuesta inicial entre el 11 y 14 de marzo.
    
    Parámetros:
    - df: DataFrame con los datos de eventos
    - orden: 'A' para ascendente, 'D' para descendente, None para sin ordenar

    Retorna:
    - DataFrame filtrado con la información solicitada
    """
    # Asegurarse de que la columna Fecha sea tipo datetime
    df['Fecha'] = pd.to_datetime(df['Fecha'])

    # Filtrado por fechas
    fecha_inicio = pd.to_datetime("2025-03-11")
    fecha_fin = pd.to_datetime("2025-03-14")
    filtro_fecha = (df['Fecha'] >= fecha_inicio) & (df['Fecha'] <= fecha_fin)

    # Filtrado por encuesta inicial y respuesta enviada
    filtro_evento = (
        df['Contexto del evento'].str.contains("Encuesta: Encuesta inicial estudiantes CDO", na=False) &
        df['Nombre evento'].str.contains("Respuesta enviada", na=False)
    )

    # Aplicar filtros
    df_filtrado = df[filtro_fecha & filtro_evento].copy()

    # Ordenar si se indica
    if orden == 'A':
        df_filtrado = df_filtrado.sort_values(by='Fecha', ascending=True)
    elif orden == 'D':
        df_filtrado = df_filtrado.sort_values(by='Fecha', ascending=False)

    return df_filtrado

    """
