# Solución
# EJERCICIO 1

```python

from pathlib import Path
import pandas as pd

file_route  = Path("archivos") / 'logs_catedra.csv'

data_frame = pd.read_csv (file_route, delimiter=',', encoding='utf-8')

```

```python
def order_data_frame (data_f):

    filter_df = data_f[(data_f["Contexto del evento"] == 'Foro: Foro armado de grupo') & (data_f['Nombre evento'] == 'Módulo de curso visto')]

    order_df = filter_df.sort_values (by='Nombre completo del usuario')
    return order_df
```

```python
data_frame_filtered = order_data_frame (data_frame)

data_frame_filtered
```

```python
users_no_repetition = data_frame_filtered["Nombre completo del usuario"].drop_duplicates()

last_top_5 = users_no_repetition.tail(5)

last_top_5
```

```python
# se interpreta que la cantidad de registros del conjunto de datos completo es la cantidad de filas
# de X usuario
result = []
for user in last_top_5:
    cant = data_frame[data_frame["Nombre completo del usuario"] == user].shape[0]
    result.append({"Nombre completo del usuario": user, "Cantidad de registros": cant})

df_result = pd.DataFrame(result)
df_result
```

```python
import pandas as pd

# EJERCICIO 1 usando query y sort_values + drop_duplicates
def obtener_registros_foro_armado():
    # Leer CSV
    df = pd.read_csv('archivos/logs_catedra.csv')

    # Filtrar usando query
    filtrado = df.query(
        "`Contexto del evento` == 'Foro: Foro armado de grupo' and `Nombre evento` == 'Módulo de curso visto'"
    )

    # Ordenar por nombre
    filtrado_ordenado = filtrado.sort_values('Nombre completo del usuario')

    return filtrado_ordenado, df  # Devuelvo también el CSV completo


# EJERCICIO 2 usando value_counts y merge
def mostrar_ultimos_cinco_usuarios():
    filtrado, df_completo = obtener_registros_foro_armado()

    # Sacar usuarios únicos ordenados alfabéticamente usando drop_duplicates + sort_values
    usuarios_ordenados = filtrado[['Nombre completo del usuario']] \
        .drop_duplicates() \
        .sort_values('Nombre completo del usuario')

    # Tomar los últimos 5 con iloc
    ultimos_cinco = usuarios_ordenados.iloc[-5:]

    # Contar cantidad de registros de cada usuario en TODO el CSV usando value_counts
    conteos = df_completo['Nombre completo del usuario'].value_counts().reset_index()
    conteos.columns = ['Nombre completo del usuario', 'Cantidad de registros']

    # Hacer merge para quedarme solo con los 5 seleccionados y sus cantidades
    resultado = ultimos_cinco.merge(conteos, on='Nombre completo del usuario', how='left')

    # Mostrar resultado
    print("\nÚltimos cinco usuarios (alfabéticamente) y cantidad de registros en TODO el CSV:\n")
    print(resultado.to_string(index=False))



mostrar_ultimos_cinco_usuarios()
```
